```python
# Yahoo.jp 
def yahoo(search, size = True):
    search_terms = search.split()
    
    # Prepare URL with search terms 
    url ="https://auctions.yahoo.co.jp/search/search?auccat=&tab_ex=commerce&ei=utf-8&aq=-1&oq=&sc_i=&fr=auc_top&p="
    
    for i in range(len(search_terms)):
        url += search_terms[i]+"+"
    if size == True:
        s=28
        #s = input("Please enter your shoe size (JP)\n (i.e US10 == 28 cm)\n:")
        url = url + str(s) + "&x=0&y=0"
    else:
        url = url + "&x=0&y=0"
        
    # Initialize and obtain html info 
    r = requests.get(url = url)
    if r.status_code == 200:
        print("Access successful")
        print("Searching for " + search)
    else:
        print("Unable to fetch.")
        sys.exit()
    uclient= urlopen(url)
    page_html=uclient.read()
    uclient.close()
    page_soup= soup(page_html, "html.parser")
    listings=page_soup.find_all("li", {"class":"Product"})
    
    if listings == None:
        print("Nothing found")
        
    else:
        for listing in listings:
            name = listing.find("img", {"class":"Product__imageData"})
            name = name.get("alt")
            print("Listing: "+ name)
            price = listing.find("span", {"class":"Product__priceValue u-textRed"}).text
            print("Current price: "+ price+"\n")
```

```python
opt = Options()
opt.add_argument("--incognito")
opt.add_argument("--headless")
opt.add_argument("--ignore-certificate-errors")
driver = WebDriver('/Users/seanlow/Desktop/2020/Vintage Scraper/chromedriver',options=opt)

# Grailed 
def grailed(search):
    def init_browser():
        url = "https://www."+ "grailed" + ".com"
        driver.get(url)
        try:
            WebDriverWait(driver, 10).until(
                lambda driver: driver.execute_script('return document.readyState') == 'complete')
        except TimeoutException:
            print("It took too much time")
            return False
        return True
    access = init_browser()
    if access:
        try: 
            WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.ID,"globalheader_search")))
        except NoSuchElementException:
            print("Searchbar not found.")

        searchbar = driver.find_element_by_id("globalheader_search")
        searchbar.send_keys(search)
        searchbar.send_keys(Keys.ENTER)
        try:
            WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CLASS_NAME, "listing-cover-photo")))
        except NoSuchElementException:
            print("No listings found")
    
        # Remove login prompt window (might not be necessary if I scrape fast enough)
        try:
            login_prompt = EC.presence_of_element_located((By.CLASS_NAME, "close"))
            WebDriverWait(driver, 20).until(login_prompt)
            exit_button = driver.find_element_by_xpath('//a[@class="close"]')
            exit_button.click()
            print("Done")
        except NoSuchElementException:
            print("No login prompt detected")
        
        listings = driver.find_elements_by_class_name("feed-item")

        for listing in listings:
            try:
                name = listing.find_element_by_class_name("listing-title").text
                print(name)
                price = listing.find_element_by_class_name("listing-price").text.split()
                if len(price) == 2:
                    current_price = price[0]
                    old_price = price[1]
                    price = "Current: "+ current_price + " Original: "+ old_price
                else: 
                    price = "Current: "+ price[0]
                
                print(price)
            except NoSuchElementException:
                print("End")
                
        print("Grailed scraping DONE")
        driver.close()    
        
        

        
```
